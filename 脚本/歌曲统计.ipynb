```python
import os
```

```python
music = {}
```

```python
path = r'E:\OneDrive365\OneDrive\音乐\【05】经典怀旧歌曲无损WAV格式合集'
files = os.listdir(path)
```

```python
for file in files:
    info = file.split('-')
    if len(info) == 2:
        key, value = info[0][:-1], info[1][1:-4]
    else:
        print(file)
        continue
    if key in music.keys():
        music[key].append(value)
        print('已有: ',file)
    else:
        music[key]=[]
        music[key].append(value)
(music)
```
