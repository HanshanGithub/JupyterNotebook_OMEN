```python
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from time import sleep
from datetime import date
from tkinter import *
from threading import Thread
import tkinter
```

```python

class Xhu():
    def __init__(self, root):
        self.urls = r'http://xhbw.xhu.edu.cn/'
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")
        #启动开发者模式，隐藏标签
        options.add_experimental_option("excludeSwitches",['enable-automation'])
        #禁止Chrome带有正在被自动化控制的信息
        options.add_argument("--disable-blink-features=AutomationControlled")
        options.add_experimental_option('mobileEmulation', {'deviceName': 'iPhone X'})  # 模拟iPhone X浏览
        self.driver = webdriver.Chrome(options=options)
        self.driver.set_window_size(500, 800)
        self.driver.get(self.urls)
        self.now_day = date.today().day

        self.xhu_id = StringVar()
        self.xhu_pwd = StringVar()

        self.initGUI(root)

    def initGUI(self, root):
        root.title("")
        root.geometry('400x300')

        l1 = Label(root, text="学号")
        l1.pack()  # 这里的side可以赋值为LEFT  RTGHT TOP  BOTTOM

        self.xhu_id_ = Entry(root, textvariable=self.xhu_id)
        self.xhu_id.set("")
        self.xhu_id_.pack()
        l2 = Label(root, text="密码")
        l2.pack()  # 这里的side可以赋值为LEFT  RTGHT TOP  BOTTOM

        self.xhu_pwd_ = Entry(root, textvariable=self.xhu_pwd)
        self.xhu_pwd.set("")
        self.xhu_pwd_.pack()
        Button(root, text="登录", command=self.log).pack()
        
        self.text = Text(root, width=50, height=10)
        scroll = tkinter.Scrollbar()
        scroll.pack(side=tkinter.RIGHT,fill=tkinter.Y)
        scroll.config(command=self.text.yview)
        self.text.config(yscrollcommand=scroll.set)
        self.text.pack()
        
        # self.text.configure(state='disabled')
        root.mainloop()

    def log(self):
        T = Thread(target=self.__log, args=())
        T.start()

    def __log(self):
        # 更改id pwd 位置信息做提示
        id_ = self.driver.find_element("xpath", "//input[@type='text']")
        id_.send_keys('3120190912213') #    self.xhu_id.get()
        sleep(0.1)
        pwd = self.driver.find_element("xpath", "//input[@type='password']")
        sleep(0.1)
        pwd.send_keys('xhdxbw0410') #    self.xhu_pwd.get()
        login = self.driver.find_element("xpath", "//button[@class='submit_but']")
        login.click()
        sleep(0.5)
        try:
            self.driver.find_element("id", "errorBox")
        except:
            self.xhu_id.set("")
            self.xhu_pwd.set("")
            self.text.insert(INSERT, "已成功登录")
            self.doing()
        else:
            self.text.insert(INSERT, "检查:" + self.xhu_id.get() + ' ' + self.xhu_pwd.get() + '\n')
            self.xhu_id.set("信息有误")
            self.xhu_pwd.set("重新输入")
            id_.clear()
            pwd.clear()
            self.driver.refresh()
            return

    def doing(self):
        self.back_to_apply()
        for d in range(self.now_day, 31):
            self.text.insert(END, '1月' + str(d) + '号到' + str(d + 1) + '号请假中...\n')
            self.select_time("startTime", d)
            self.select_time("endTime", d + 1)
            self.else_to_ok()
            self.withdraw()
            self.back()
            self.text.insert(END, '1月' + str(d) + '号到' + str(d + 1) + '号已请假完成\n')
            self.text.insert(END, '\n')
        self.quit_xhu()

    def back_to_apply(self):
        vacation = self.driver.find_element("xpath", "//img[@src='/images/wechat/leave_apply.png']")
        vacation.click()
        sleep(1)

    def select_time(self, date_index, date):
        time_select = self.driver.find_element("id", date_index)  # endTime startTime
        time_select.click()
        sleep(0.5)
        now_d = self.driver.find_element("xpath", "//div[@data-id='picker-d']/div/ul/li[@class='highlight']").text
        day = int(now_d)
        li_d = self.driver.find_elements("xpath", "//div[@data-id='picker-d']/div/ul/li")
        if date_index == "startTime":
            for i in range(day, date):
                sleep(0.1)
                li_d[i].click()

        if date_index == "endTime":
            now_i = self.driver.find_element("xpath", "//div[@data-id='picker-i']/div/ul/li[@class='highlight']").text
            li_i = self.driver.find_elements("xpath", "//div[@data-id='picker-i']/div/ul/li")
            for i in range(day, date):
                sleep(0.1)
                li_d[i].click()
            sleep(1)
            li_i[int(now_i) - 1].click()
        self.date_ok()

    def date_ok(self):
        but_ok = self.driver.find_element("xpath", "//button[@data-id='btn-ok']")
        but_ok.click()
        sleep(1)

    def else_to_ok(self):
        dination = self.driver.find_element("xpath", "//textarea[@placeholder='请输入目的地']")
        dination.send_keys('.')
        reason = self.driver.find_element("xpath", "//textarea[@placeholder='请注明目的地（具体到区县）']")
        reason.send_keys('.')
        go_area = self.driver.find_element("xpath", "//div[@class='el-input el-input--suffix']")
        go_area.click()
        cd = self.driver.find_elements("xpath", "//div[@class='el-scrollbar']/div/ul/li")[-1]
        sleep(0.5)
        cd.click()
        sleep(0.5)
        over = self.driver.find_element("xpath", "//div[@class='sbmit_btn']")
        over.click()
        sleep(1)

    def withdraw(self):
        sleep(3)
        # //*[@id="application"]/div[1]/ul/div[4]
        rediraw = self.driver.find_element("xpath", "//*[@id='application']/div[1]/ul/div[4]")
        rediraw.click()
        sleep(1)
        # /html/body/div[2]/div[2]/span[2]
        cancel = self.driver.find_element("xpath", "/html/body/div[2]/div[2]/span[2]")
        cancel.click()

    def back(self):
        self.driver.back()
        self.driver.refresh()
        sleep(1)
        # 请到31号完毕退出TK()

    def quit_xhu(self):
        self.xhu_id.set("已请假完毕")
        self.xhu_pwd.set("程序将在5秒后退出")
        self.driver.quit()
```
